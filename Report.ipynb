# 1. Анализ датасета
#### Посмотрим подготовленный сет для обучения неронной сети, данный сет прдназначен для детекции авто, как стоящих, так и в движение. Данный датасет является типичным форматом для сетки yolo, который мы и воспользуемся

##### Кратко про алгоритм YOLO(You only look once). Алгоритм на данный момент времени является наверное лучшим решением для машинного зрения в задачах: 
- ##### Детекции
- ##### Сегментации 
- ##### Классификации
##### Сравним его с подобными ему алгоритмами: Faster RCNN, Mask RCNN, но уступающими в производительности. Yolo является алгоритмом на основе сеток RCNN(Region-based convulation neural networks). Faster и Mask rcnn`ы почти абсолютно похожие алгоритмы за исключением, что сетка mask еще и накладывает маски на объекты, после запоминая их, тем самым способствуя лучшему запоминанию. 

##### Yolo является более легковесным алгоритмом, улучшая время предикта, но при этом почти не теряя качество метрик. Исходя из всего этого был выбран алгоритм Yolo от компании ultralytics 8-й версии - YOLOv8

# 2. Создадим файл data.yaml для передачи в него основной информации о нашем датасете
##### Сохраняя структуру изначально данных классов передадим такие данные в созданный файл: путь до train, val выборок; количество классов; имена классов(в неком массиве через запятую).

##### Также необходимо изменить путь до нашего датасета в файле от ultralytics - "settings.yaml", который находится по пути: AppData/Roaming/Ultralytics/settings.yaml
